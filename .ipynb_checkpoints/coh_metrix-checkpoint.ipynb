```python
def submission_count(filename):
    fullname= '/home/bansal/Downloads/' + filename + '.json'
    print (fullname)
    try:
        count= 0
        with open(fullname, 'r') as data_file:
            data= json.load(data_file)
            for post in data:
                if(post['link_id']== post['parent_id']):
                    count+= 1
            return count
    except Exception as e:
        print (e)
        return 0
```
